# 好奇
## 帖城教会是一个怎么样的教会
最近《2031》刚学完帖撒罗尼迦前后书，其中有个非常惊奇的点，就是保罗只在帖城待了很短的时间(大概1个月？参见使徒行传17章)。
这么短的时间能干什么呢？

如果是我们在建立教会，只在一个城市待了一个月的时间就被迫离开，那么接下来这个教会会是什么样的光景呢？

所以想先重点从使徒行传和帖撒罗尼迦书信中先查看帖城的教会在保罗离开前后的光景是如何？

帖城弟兄姐妹的特点和经历：
- 不少人被说服，其中很多是敬畏上帝的希腊人和有地位的妇女
- 接待保罗，住在自己家中
- 深受苦难
- 效法使徒和耶稣、喜乐领受，成为榜样
- 传道传福音到马其顿和亚该亚以及其他各地
- 离弃偶像归向上帝、等候主再来
- 在各种迫害和患难中仍然存坚忍的信心

这样一个从来没有听过福音，只被牧者带领了1个月左右的一群人，居然在带领者被迫离开后没有解散，反而还愈发刚强和成长。
这太不可思议了。

如果把前面的文字捂住，只看后面这些特点，问一个人说你认为一个教会要达到上面的光景需要多少时间？
恐怕没有人敢说几个月就可以。

但是帖城的弟兄姐妹们做到了，而且他们是在两千年前做到的，没有一本完整的圣经，感觉在知识上也应该没有我们这群人了解的多。
不管是圣经真理知识，还是世上的文字语言或者其他科学知识。

我相信他们肯定很喜欢常常聚在一起，常常掰饼聚会祷告，不然肯定达不到上面那些特点。
不知道他们每次聚会内容是什么？会有人带领他们吗？每次聚会分享什么？学习什么？他们觉得每次聚会有意义吗？
应该不是只是去打个卡吧。那么是什么在吸引他们呢？

是什么让他们觉得这么做是非常有意义和有价值呢？甚至遭遇逼迫也在所不惜！

反观我们这些已经是对上帝对福音有一定的认识，和神有一定经历的人，都没有达到这样的光景。
而且我们手上有各种版本的圣经和非常多的人的研经资料，都常常觉得聚会不够吸引我们。
如果我们能时间穿越的话，你想象中的帖城弟兄姐妹们的聚会情景是怎么样的呢？
假设你现在正站在帖城教会聚会的门外，你觉得推开门后会看到一群什么样的人呢？

既然从外面看帖城的弟兄姊妹没有什么地方是比我们更强的，那结的果子比我们多的唯一原因就来自里面了。
他们一定是在实际做工中蒙了很多神的带领，常常被圣灵感动，
所以每当他们聚在一起的时候就是在分享各自被圣灵带领的地方以及他们里面的感动。

他们一定是彼此相爱的，而且爱就是他们之间唯一的关系。
他们一起做工，一起唱诗，一起祷告，一起赞美！

如果现在世上就有这么一个教会，我想很多人是非常感兴趣想要去参观和学习的，
那么现在这么好的机会摆在面前，我们不妨好好再看看帖城教会。
到底保罗在他们中间时是怎么教导他们的？
他们是怎么回应教导的？
圣灵在帖城工作的痕迹是什么，在他们身上留下了什么样的印记？
而且保罗被迫离开后，当有机会再写信时会提到哪些真实的经历，重点要嘱咐的是什么？
