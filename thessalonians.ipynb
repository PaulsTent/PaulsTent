最近《2031》刚学完帖撒罗尼迦前后书，其中有个非常惊奇的点，就是保罗只在帖城待了很短的时间(大概1个月？参见使徒行传17章)。

如果是我们在建立教会，只在一个城市待了一个月的时间就被迫离开，那么接下来这个教会会是什么样的光景呢？

所以想先重点从使徒行传和帖撒罗尼迦书信中先查看帖城的教会在保罗离开前后的光景是如何？

帖城弟兄姐妹的特点和经历：
- 不少人被说服，其中很多是敬畏上帝的希腊人和有地位的妇女
- 接待保罗，住在自己家中
- 深受苦难（来自同胞和犹太人的逼迫）
- 效法使徒和主耶稣、喜乐领受，成为榜样
- 传道传福音到马其顿和亚该亚以及其他各地
- 离弃偶像归向上帝、等候主再来
- 在各种迫害和患难中仍然存坚忍的信心
- 确信这不是人的道理，而是神的道
- 自己从上帝那里领受了要彼此相爱

这样一个从来没有听过福音，只被牧者带领了1个月左右的一群人，居然在带领者被迫离开后没有解散，反而还愈发刚强和成长。
这太不可思议了。

如果把前面的文字捂住，只看后面这些特点，问一个人说你认为一个教会要达到上面的光景需要多少时间？
恐怕没有人敢说几个月就可以。

为了更好的看见保罗对帖城的弟兄姐妹的负担，接下来会展示帖撒罗尼迦前后书的词云，
看看哪些词是重复出现的。

```python
from os import path
import jieba
from wordcloud import WordCloud, STOPWORDS

d = path.abspath('')
text = open(path.join(d, 'text/帖撒罗尼迦前书.txt')).read()
seg_text = " ".join(jieba.cut(text))
font_path='/System/Library/Fonts/PingFang.ttc'
stop_words = set(STOPWORDS)
stop_words.update(set('你们我们的神弟兄们'))
stop_words.add('你们')
word_cloud = WordCloud(
    background_color="white",# 设置背景颜色
    max_words=100, # 词云显示的最大词数
    stopwords=set(STOPWORDS), # 设置停用词
    font_path=font_path, # 兼容中文字体，不然中文会显示乱码
).generate(seg_text)
%pylab inline
import matplotlib.pyplot as plt
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis("off")
```

## 保罗在帖城的时候：

保罗的做事、经历和态度：
- 按照以往的习惯进入会堂（*问题：我们的习惯是什么？传福音的切入点是什么？*）
- 一连三个安息日引用圣经解释证明基督的受害和复活(和众人一起查考圣经)，而且是**放胆**传扬
   - 传福音不只靠言语也靠上帝的能力、圣灵的同在和充分的确据
- 住在耶孙的家里
- 常常为帖城肢体感谢上帝，并为他们祷告
- 行事为人：一面传福音一面辛勤工作；
- 目的不是取悦人而是取悦上帝，没有花言巧语和贪念
- 像母亲抚育，甚至**舍命也在所不惜**；父亲般的的教导；
- 看待帖城弟兄姐妹就是自己的荣耀和喜乐

遇到的环境：
- 有不少人被说服成为同道
- 不信主的犹太人召集无赖引发骚乱闯进耶孙的家企图迫害保罗和西拉，迫使保罗离开帖城
- 百姓和当地官长对不信主犹太人说的另有一个王叫耶稣深感不安
- 帖撒罗尼迦人不如庇哩亚人通情达理
- 帖撒罗尼迦人赶到庇哩亚去怂恿煽动众人
- 撒旦拦阻保罗回到帖城

保罗的教导：
- 预言自己将会必受患难
- 彼此相爱以及对他人的爱满溢出来
- 更加努力：圣洁远避淫乱，持守身体的圣洁和尊贵
- 安分守己亲手做工
- 在主里的安息与盼望主再来时的复活
- 主再来时像贼在夜里忽然出现一样
- 做光明之子，保持清醒：信望爱
- 敬重辛勤工作、在主里领导和劝诫你们的人
