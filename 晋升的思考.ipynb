## 晋升看重什么
大部分晋升是有答辩流程的，不管这个流程是否是走个过场。
其中最常被问到的三个问题是：

- 技术选型
- 周边架构
- 效果数据

### 技术选型的问答
例如为什么要用A这个技术，而不是B或者C技术。最错误的答案是：我也不知道，领导让我用的。

正确一点应该是对各个技术的综合各方比较，列一个表出来，最后阐明选择的原因。

### 周边架构的问答
考察你对业务整体的理解。和你相关，但并不是你日常编码所实现的工作。
比如你是做后端开发的，存储可能是你们公司的基础设施，你直接调用就行。
但评委很有可能会问你的存储是如何做容灾、备份等保护措施的。如果回答不知道，照样完蛋。

总的来说在P5 和 P6 的晋升上只要能把自己负责的模块实现了，
在性能上做 1 - 2 个提升就差不太多了。
但是到了 P7 再往上，更多的是考验候选人的对全局把握的能力。
上面这三点其实用一个词来概括就是架构，
考察业务全局、公司、乃至业界等更大视角的理解和把握能力。

### 效果数据的问答
你所做的工作在效果上有多大的提升，和业界同类的技术相比如何。
如果你完全没思考过，想回答上来也很难。
而且这个效果不止体现在个人工作上的局部效果，而是对公司整体业务的影响和提升是什么。

现在大厂分工越来越细。你日常工作所编写的代码往往只是整个公司的总架构下的一环。
如果你能够理解和看懂公司整体架构的话，工作目标更清晰，开发起来也更得心应手，
甚至有的时候还能达到四两拨千斤的效果。

从另外一个角度讲，分工变的很细以后，大厂也更需要有人能对公司的整体架构体系有理解的人，
而不真的希望你只是拘泥于完成自己手头模块的螺丝钉。
这也是为何总爱问你所开发模块周边系统的原因。

这篇文章显然不想深究程序员应该怎么在岗位上晋升，目的是想借此探讨下在属灵之路上的历程。

### 教会和服侍

以上面为例，我也问我自己三个问题，思考下我应该怎么回答。

#### 问题1：你为什么做1号和麦子事工？
有一个可能是比较差的回答：牧师呼召大家做这个的。其他弟兄姊妹也比较热心做这个事工。

首先需要承认的是，确实当前这个事工已经是教会摸索了好多年甚至是几十年综合下来得到的神的带领的方向。
自己既然蒙神的呼召来到这个教会，自己也想要委身于教会，那么看起来有份于教会最看重的事工是个理所当然的事。

我们拥抱教会拥抱她的事工愿意、传承使命的DNA是没有错的，但是当有机会的时候，还是想再思考下，为什么是这个事工。
（我是不是假模假样的感动自己让自己对1号和麦子有怜悯，但是却对自己的家人和邻居毫无感觉？）

首先自己是否有感动，就我自己而言，里面不见得一下子对这个事工这么有感动，对1号和麦子有那么大的怜悯和信心。
但是这两个事工都有关主耶稣所托付的大使命：传福音。

最近越来越发现一个事实：不传福音就不知道福音为何物，不传福音就不知道福音的大能或者说神的大能为何物。
按照我之前的理解，一个人想要传授一个知识，首先是自己掌握这个知识，然后才能教导给别人。
而且自己掌握的越好，教授的效果也就越好。
就好像某一门课程似的，有的人不喜欢学是因为老师没有掌握好，没有教出精华来。
但是福音好像不是这样的，我们越是想先掌握福音再去传讲福音，越是发现自己掌握不了福音。
我似乎突然明白，其实福音不是让人掌握的，福音本身就是一种随处随在的感动和信心。
他不是客观的一种事物，而像是主观存在的，他或在你心中或不在你心中。（写到这里我多么盼望自己能被福音充满，
被福音的大能所碾压和更新。）

福音他自己是自由的，他会出现在那些里面真心火热，对人有怜悯有爱心，是耶稣基督忠心的仆人心里面。
他会出现在那些真心痛悔，认识到自己是个彻头彻尾的罪人心里面。（第一次遇见福音的时候，不是感叹怎么获得福音，
而是惊叹居然我里面有这福音！）

所以福音绝对不是事工，而是神的大能和奥秘，福音也绝对不是知识，要搞出一套极其晦涩难懂的理论出来，
而是来自神的极大怜悯和极深的恩典。

为何一个人可以被用尽，为何一个罪人悔改之后能产生如此大的能量，是因为耶稣。
当跑的路跑尽的时候就是这个人完全没有一丝保留，被主用尽的时候。
主啊愿我的生命每一丝都能被使用来诉说和传讲你的福音——耶稣基督并你所钉十字架。
