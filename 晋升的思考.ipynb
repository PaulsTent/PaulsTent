最近看了一篇公众号讲到晋升答辩的事情，有所感触想动笔记录一下，期望能有圣灵的带领~

为什么想写这个内容呢，并不是因为最近我个人得到了晋升之类的，而是因为晋升这个事情是大家比较看重的，那么为了晋升我们需要做好哪些事情呢？
在属灵的生命中，又有哪些是可以和这个相对应，有哪些是关乎我们属灵生命成长的呢？

背景：软件工程师在大小厂中的晋升之路，虽然我个人并不擅长这块，对此也是很平庸普通，
但是想借此举一反三能思考一些神的话语。

## 晋升看重什么
大部分晋升是有答辩流程的，不管这个流程是否是走个过场。
其中最常被问到的三个问题是：

- 技术选型
- 周边架构
- 效果数据

### 技术选型的问答
例如为什么要用A这个技术，而不是B或者C技术。最错误的答案是：我也不知道，领导让我用的。

正确一点应该是对各个技术的综合各方比较，列一个表出来，最后阐明选择的原因。

### 周边架构的问答
考察你对业务整体的理解。和你相关，但并不是你日常编码所实现的工作。
比如你是做后端开发的，存储可能是你们公司的基础设施，你直接调用就行。
但评委很有可能会问你的存储是如何做容灾、备份等保护措施的。如果回答不知道，照样完蛋。

总的来说在P5 和 P6 的晋升上只要能把自己负责的模块实现了，
在性能上做 1 - 2 个提升就差不太多了。
但是到了 P7 再往上，更多的是考验候选人的对全局把握的能力。
上面这三点其实用一个词来概括就是架构，
考察业务全局、公司、乃至业界等更大视角的理解和把握能力。

### 效果数据的问答
你所做的工作在效果上有多大的提升，和业界同类的技术相比如何。
如果你完全没思考过，想回答上来也很难。
而且这个效果不止体现在个人工作上的局部效果，而是对公司整体业务的影响和提升是什么。

现在大厂分工越来越细。你日常工作所编写的代码往往只是整个公司的总架构下的一环。
如果你能够理解和看懂公司整体架构的话，工作目标更清晰，开发起来也更得心应手，
甚至有的时候还能达到四两拨千斤的效果。

从另外一个角度讲，分工变的很细以后，大厂也更需要有人能对公司的整体架构体系有理解的人，
而不真的希望你只是拘泥于完成自己手头模块的螺丝钉。
这也是为何总爱问你所开发模块周边系统的原因。

这篇文章显然不想深究程序员应该怎么在岗位上晋升，目的是想借此探讨下在属灵之路上的历程。

### 教会和服侍

以上面为例，我也问我自己三个问题，思考下我应该怎么回答。

#### 问题1：你为什么做1号和麦子事工？
有一个可能是比较差的回答：牧师呼召大家做这个的。其他弟兄姊妹也比较热心做这个事工。

首先需要承认的是，确实当前这个事工已经是教会摸索了好多年甚至是几十年综合下来得到的神的带领的方向。
自己既然蒙神的呼召来到这个教会，自己也想要委身于教会，那么看起来有份于教会最看重的事工是个理所当然的事。

我们拥抱教会，拥抱她的事工，愿意传承使命的DNA是没有错的，但是当有机会的时候，还是想再思考下，为什么是这个事工。
（我是不是假模假样的感动自己让自己对1号和麦子有怜悯，但是却对自己的家人和邻居毫无感觉？）

首先自己是否有感动，就我自己而言，里面不见得一下子对这个事工这么有感动，对1号和麦子有那么大的怜悯和信心。
但是这两个事工都有关主耶稣所托付的大使命：传福音。

最近越来越发现一个事实：不传福音就不知道福音为何物，不传福音就不知道福音的大能或者说神的大能为何物。
按照我之前的理解，一个人想要传授一个知识，首先是自己掌握这个知识，然后才能教导给别人。
而且自己掌握的越好，教授的效果也就越好。
就好像某一门课程似的，有的人不喜欢学是因为老师没有掌握好，没有教出精华来。
但是福音好像不是这样的，我们越是想先掌握福音再去传讲福音，越是发现自己掌握不了福音。
我似乎突然明白，其实福音不是让人掌握的，福音本身就是一种随处随在的感动和信心。
他不是客观的一种事物，而像是主观存在的，他或在你心中或不在你心中。（写到这里我多么盼望自己能被福音充满，
被福音的大能所碾压和更新。）

福音他自己是自由的，他会出现在那些里面真心火热，对人有怜悯有爱心，是耶稣基督忠心的仆人心里面。
他会出现在那些真心痛悔，认识到自己是个彻头彻尾的罪人心里面。（第一次遇见福音的时候，不是感叹怎么获得福音，
而是惊叹居然我里面有这福音！）

所以福音绝对不是事工，而是神的大能和奥秘，福音也绝对不是知识，要搞出一套极其晦涩难懂的理论出来，
而是来自神的极大怜悯和极深的恩典。

为何一个人可以被用尽，为何一个罪人悔改之后能产生如此大的能量，是因为耶稣。
当跑的路跑尽的时候就是这个人完全没有一丝保留，被主用尽的时候。
主啊愿我的生命每一丝都能被使用来诉说和传讲你的福音——耶稣基督并你所钉十字架。

所以回过头来，再看这个问题，为什么是1号和麦子事工？我想是因为福音的缘故，我欠福音的债吧~

#### 问题2：你觉得教会需要一个人在哪些方面的成长才能更好的建造教会？
首先确实对于这个问题我也给不出一个特别合适全面的答案，这里只想借着一个点展开表达下。
我们知道保罗在布道的时候很多次提到他是一边织帐篷维持生活所需，一边传福音。
如果我们现在也学习保罗，一边织帐篷一边传福音会怎么样呢？我想会卖不出去帐篷吧。

为什么我会这么认为呢，因为我觉得我们自己不专业织出来的肯定不如淘宝上的好看，估计质量也不行，而且成本还很贵。
淘宝上的都是批量采购原材料，机械的流水线式生产，和顶级的老师傅比起来会有差距，但是和专业是传道的牧师比起来肯定是好不少的，而且价格便宜。
现今社会是一个分工特别细致，流程特别复杂的生产方式，这样的好处是把一个产品每个环节都拆解，让每个零部件和过程都打磨到极致。
我们现今的人也是在这样的环境中工作，我们像螺丝钉一样的在工作岗位中工作，我们的焦点和目光往往都在特别细节和局部的事情上。
站在一个公司运营的角度来看，老板除了希望员工能兢兢业业之外，最希望的是公司的产品有创新，有持续的竞争力。
这种创新往往来自于对整体的把控，老板自己其实也不是特别明确公司未来几年的规划是什么，还应该拓宽什么市场，他也希望有人能给出更好的建议。

回到教会当中，我们应该不满足当下的聚会，不满足自己的事工，还应该有基督的身体观。
求主拓宽我们的心，神对教会的心意是怎样的，什么样才叫丰满。是否对别的城市的教会特别有亲情感，是否对别的事工有特别的需要感。
我们所做的事工只是神荣耀计划的一小部分，当全部事工都完成时，整个画面展开后会是多么的宏伟壮丽。

我想说的是，陷入到局部往往会找不清方向，放眼全局，从更大的视角来看，会让我们更加有信心，有盼望。

#### 问题3：你的服侍果效如何？
在公司里，非常不鼓励用感性的词语来表达自己工作的结果，期望看到的是量化后的具体数字指标有哪些变化。
为什么我们期望看到数字，因为觉得数字是客观的，是可以被掌控的（至少是期待掌控，太感性了就不知道怎么掌控）。
机器怎么理解世界，是依赖对任务目标进行的数字化的表示。世界也想让我们成为没有情感的人，也想让我们以数字的眼光看这个世界，
当我们看到一件事的时候，脑子里是一堆数字，看到人的时候脑子里也是一堆数字。都是一些冷冰冰可以互相比较的数字。
在公司里我们觉得一个人专业，就是他能够很好的量化自己要做的事，清晰地知道做哪些事可以对公司的业务有些量化的影响。
这样的人，做事情很有章理，也很标准很严谨。
他可能能给公司带来很高的价值，同时他的收入也高，可能周围的人都赞许他，但是有一个很感恩的事就是，如果他有一个刚会走路还不太会说话的孩子，
这个孩子可能不会特别的爱这样的爸爸，因为觉得他太冷漠了，没有神的形象和样式。

再想到自己选择工作的行业，觉得在体制内上班太不好了，各种勾心斗角，尔虞我诈。当个医生也不好，收受红包，
因为业绩的压力给治疗的家庭开很多无用的检查和医药。当律师也不好，很多时候不能坚持法律和公义。
思来想去觉得还是当个程序员好，天天跟电脑打交道，总不需要勾心斗角，昧着良心做事，而且电脑程序只接受“公义”
的程序才能正确运行。但是没想到罪有千千万万款，总有一款适合你，这样的工作让我本来就稀缺的爱变得越来越少了。

再谈到服侍的效果，我们可以用数字来描述吗？可以用，但是要小心，因为我们不是机器，不是为了数字活着的，而是有特殊的身份：
唯有我们是被拣选的族类，是有君尊的祭司，是圣洁的国度，是属神的子民。。。
